```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
def FF(HL,OL,W,B,x):
  INL=len(x)
  hidden_layer=np.zeros((HL))
  output_layer=np.zeros((OL))
  witr=0
  bitr=0
  for i in range(HL):
    sum=0
    for j in range(INL):
      sum=sum+(x[j]*W[witr])
      witr=witr+1
    sum=sum.sum()
    hidden_layer[i]=sigmoid(sum+B[bitr])
    bitr=bitr+1

  for i in range(OL):
    output=0
    for j in range(HL):
      output=output+(hidden_layer[j]*W[witr])
      witr = witr + 1
    output=output.sum()
    output_layer[i]=sigmoid(output+B[bitr])
    bitr=bitr+1
  return output_layer
```

```python
def POA(prob,lb,up,dim,N,iters):

  if not isinstance(lb,list):
    lb=[lb]*dim
  if not isinstance(up,list):
    up=[up]*dim
  
#Initialization
  X=np.zeros((N,dim))
  fit=np.zeros(N)
  for i in range(dim):
    X[:,i]=lb[i]+np.random.uniform(0, 1, N) * (up[i] - lb[i])

  for i in range(N):
    L=X[i,:]
    fit[i]=prob(L)

  for t in range(iters):
    best=min(fit)
    location=np.argmin(fit,axis=0)
    if t==0:
      Xbest=X[location,:]
      fbest=best
    elif best<fbest:
      fbest=best
      Xbest=X[location,:]
#Update location of food

    X_FOOD=[]
    k=random.randint(0,N-1)
    X_FOOD=X[k,:]
    F_FOOD=fit[k]

    for i in range(N):
      #Phase 1 Moving towards prey(exploration_phase)
      I=round((1+np.random.uniform(0, 1, 1))[0])
      if fit[i]>F_FOOD:
        X_new=X[i,:]+(np.random.uniform(0, 1, 1)*(X_FOOD-(I*X[i,:])))
      else:
        X_new=X[i,:]+(np.random.uniform(0,1,1)*(X[i,:]-(1*X_FOOD)))
      
      #Updating X_i using 5
      f_new=prob(X_new)
      if f_new<=fit[i]:
        X[i,:]=X_new
        fit[i]=f_new
      #Phase 2
      X_new=X[i,:]+((0.2*(1-(t/iters)))*((2*np.random.uniform(0, 1, dim))-1)*X[i,:])
      
      #Updating X_i using 7
      f_new=prob(X_new)
      if f_new<=fit[i]:
        X[i,:]=X_new
        fit[i]=f_new

  Best_score=fbest
  Best_pos=Xbest  
  print(Best_score)
  print(Best_pos)
  return Best_score,Best_pos




```
